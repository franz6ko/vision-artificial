```python
# Implementación de pasaje a coordenadas cromáticas

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img1 = cv.imread('img/CoordCrom_1.png')
img2 = cv.imread('img/CoordCrom_2.png')
img3 = cv.imread('img/CoordCrom_3.png')

img1 = cv.cvtColor(img1,cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(img2,cv.COLOR_BGR2RGB)
img3 = cv.cvtColor(img3,cv.COLOR_BGR2RGB)

def crom_coord(img_in):
    rgb = img_in.sum(axis=2)[:,:,np.newaxis]
    rgb[rgb == 0] = 1
    img = (255.0*img_in)/rgb
    img = img.round()
    np.clip(img, 0, 255, out=img)
    img = img.astype('uint8')
    return img
    
res1 = crom_coord(img1)
res2 = crom_coord(img2)
res3 = crom_coord(img3)

fig = plt.figure(figsize=(8, 8), dpi=130)
ax1=plt.subplot(331)
ax1.imshow(img1)
ax2=plt.subplot(332)
ax2.imshow(img2)
ax3=plt.subplot(333)
ax3.imshow(img3)
ax4=plt.subplot(334)
ax4.imshow(res1)
ax5=plt.subplot(335)
ax5.imshow(res2)
ax6=plt.subplot(336)
ax6.imshow(res3)

print("OK")
```

```python
# Implementación del algoritmo White-Patch

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img1 = cv.imread('img/test_blue.png')
img2 = cv.imread('img/test_green.png')
img3 = cv.imread('img/test_red.png')

img1 = cv.cvtColor(img1,cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(img2,cv.COLOR_BGR2RGB)
img3 = cv.cvtColor(img3,cv.COLOR_BGR2RGB)

def white_patch(img_in):
    maxRGB = img_in.max(axis=(0,1))
    img = (255.0/maxRGB)*img_in
    img = img.round()
    np.clip(img, 0, 255, out=img)
    img = img.astype('uint8')

    return img

res1 = white_patch(img1)
res2 = white_patch(img2)
res3 = white_patch(img3)

fig = plt.figure(figsize=(8, 4), dpi=130)
ax1=plt.subplot(331)
ax1.imshow(img1)
ax2=plt.subplot(332)
ax2.imshow(img2)
ax3=plt.subplot(333)
ax3.imshow(img3)
ax4=plt.subplot(334)
ax4.imshow(res1)
ax5=plt.subplot(335)
ax5.imshow(res2)
ax6=plt.subplot(336)
ax6.imshow(res3)

print("OK")
```
