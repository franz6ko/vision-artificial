```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
titoCC_imgs = ('CoordCrom_1','CoordCrom_2','CoordCrom_3')
hands_imgs  = ('test_blue','test_green','test_red')
titoWP_imgs = ('WP_B','WP_O','WP_R')

def load_imgs(filenames):
    img1 = cv.imread('img/' + filenames[0] + '.png')
    img2 = cv.imread('img/' + filenames[1] + '.png')
    img3 = cv.imread('img/' + filenames[2] + '.png')

    img1 = cv.cvtColor(img1,cv.COLOR_BGR2RGB)
    img2 = cv.cvtColor(img2,cv.COLOR_BGR2RGB)
    img3 = cv.cvtColor(img3,cv.COLOR_BGR2RGB)
    
    return [img1,img2,img3]
```

```python
def plot_results(img, res, fsize=(8, 8)):
    fig = plt.figure(figsize=fsize, dpi=130)
    ax1=plt.subplot(331)
    ax1.imshow(img[0])
    ax2=plt.subplot(332)
    ax2.imshow(img[1])
    ax3=plt.subplot(333)
    ax3.imshow(img[2])
    ax4=plt.subplot(334)
    ax4.imshow(res[0])
    ax5=plt.subplot(335)
    ax5.imshow(res[1])
    ax6=plt.subplot(336)
    ax6.imshow(res[2])
```

```python
# Implementación de pasaje a Coordenadas Cromáticas

def crom_coord(img_in):
    rgb = img_in.sum(axis=2)[:,:,np.newaxis]
    rgb[rgb == 0] = 1
    img = (255.0*img_in)/rgb
    
    img = img.round()
    np.clip(img, 0, 255, out=img)
    img = img.astype('uint8')
    
    return img
```

```python
# Test de pasaje a Coordenadas Cromáticas

imgs = load_imgs(titoCC_imgs)
res = [crom_coord(img) for img in imgs]
plot_results(imgs, res)
```

```python
# Implementación del algoritmo White-Patch

def white_patch(img_in):
    maxRGB = np.max(img_in, axis=(0,1))
    img = (255.0/maxRGB)*img_in
    
    # Mostrar en la imagen original de donde se sacan los valores
    r = 3
    R, G, B = cv.split(img_in)
    Rx, Ry = np.unravel_index(R.argmax(), R.shape)
    cv.rectangle(img_in, (Rx-r,Ry-r), (Rx+r,Ry+r), (255,0,0), 1)
    Gx, Gy = np.unravel_index(G.argmax(), G.shape)
    cv.rectangle(img_in, (Gx-r,Gy-r), (Gx+r,Gy+r), (0,255,0), 1)
    Bx, By = np.unravel_index(B.argmax(), B.shape)
    cv.rectangle(img_in, (Bx-r,By-r), (Bx+r,By+r), (0,0,255), 1)

    img = img.round()
    np.clip(img, 0, 255, out=img)
    img = img.astype('uint8')

    return img
```

```python
# Test del algoritmo White-Patch (hands)

imgs = load_imgs(hands_imgs)
res = [white_patch(img) for img in imgs]
plot_results(imgs, res, (8, 4))
```

```python
# Test del algoritmo White-Patch (Tito)

imgs = load_imgs(titoWP_imgs)
res = [white_patch(img) for img in imgs]
plot_results(imgs, res)

# Imagenes muy saturadas con poca intensidad dan malos resultados con el white-patch
```

```python
# Implementación del algoritmo White-Patch utilizando percentiles

def white_patch_perc(img_in):
    pseudo_white = np.percentile(img_in, 50, axis=(0,1))
    pseudo_white[pseudo_white == 0] = 1
    img = (255.0/pseudo_white)*img_in
    
    img = img.round()
    np.clip(img, 0, 255, out=img)
    img = img.astype('uint8')

    return img
```

```python
# Test del algoritmo White-Patch utilizando percentiles (Tito)

imgs = load_imgs(titoWP_imgs)
res = [white_patch_perc(img) for img in imgs]
plot_results(imgs, res)
```
